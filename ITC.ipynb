```python
import pandas as pd

def rename_unnamed_columns(df):
    col_counter = 1

    new_columns = []
    for col in df.columns:
        if 'Unnamed' in col:
            new_columns.append(f'{col_counter}')
            col_counter += 1
        else:
            new_columns.append(col)

    # Asignamos los nuevos nombres de columna al DataFrame
    df.columns = new_columns
    return df


df = pd.read_csv("/content/Melchora.csv", delimiter=';')

df = rename_unnamed_columns(df)


df
```

## Descripción del algoritmo

Este algoritmo toma una lista de valores, donde cada valor es una potencia de 2 o una combinación de potencias de 2. El objetivo es llenar un `DataFrame` donde cada columna corresponde a un valor de la lista, y cada fila dentro de esa columna se llena con `'ok'` en función de las potencias de 2 que componen ese valor.

### Lógica del algoritmo

1. **Lista de valores**:
   La lista de entrada contiene valores como `2`, `4`, `8`, `16`, o combinaciones de estos, como `20 = 16 + 4`. Estos valores serán utilizados para llenar las columnas del `DataFrame`.

2. **Descomposición en potencias de 2**:
   Cada valor de la lista se descompone en las potencias de 2 más grandes posibles. Por ejemplo:
   - Si el valor es `20`, se descompone como `16 + 4`.
   - Si el valor es `8`, no necesita descomposición ya que es una potencia de 2.

3. **Asignación de las potencias a las filas**:
   Una vez descompuesto, cada potencia de 2 indica qué fila debe ser llenada con `'ok'` en el `DataFrame`:
   - Si el valor contiene `2`, se llena la **fila 1**.
   - Si el valor contiene `4`, se llena la **fila 2**.
   - Si el valor contiene `8`, se llena la **fila 3**.
   - Si el valor contiene `16`, se llena la **fila 4**.

4. **Llenado del `DataFrame`**:
   - El `DataFrame` tiene 4 filas, representando las fases (1, 2, 3, 4).
   - Cada columna corresponde a un valor de la lista.
   - Para cada valor en la lista, el algoritmo descompone el valor en potencias de 2 y llena las filas correspondientes en la columna asignada.

### Ejemplo

- **Entrada**:
  - Lista de valores: `[2, 2, 4, 4, 16, 8, 20]`
  - Cada valor se procesa de la siguiente manera:
    - `2` → Fila 1
    - `4` → Fila 2
    - `16` → Fila 4
    - `8` → Fila 3
    - `20` → Fila 2 y Fila 4 (descompuesto en `16 + 4`)

- **Salida esperada**:
  El `DataFrame` resultante tendrá las siguientes filas llenas con `'ok'`:

| fase | A+  | A-  | A+G | A-G | O>E | E>O | PS  |
|------|-----|-----|-----|-----|-----|-----|-----|
| 1    | ok  | ok  | None| None| None| None| ok  |
| 2    | None| None| ok  | ok  | None| None| ok  |
| 3    | None| None| None| None| ok  | ok  | None|
| 4    | None| None| None| None| ok  | None| ok  |

```python
fase_vals = [1, 2, 3, 4]

grupo_vals = df2['Nombre'].values

df5 = pd.DataFrame({'fase': fase_vals})


for grupo in grupo_vals:
    df5[grupo] = pd.Series([None] * len(fase_vals))  # Inicializamos las columnas con valores vacíos


df5
```

```python
segunda_columna_lista = list(map(int, segunda_columna_lista))

segunda_columna_lista
```

```python
# Definir las potencias de 2 que vamos a usar
potencias = [16, 8, 4, 2]

# Función para descomponer el valor en potencias de 2 más grandes posibles
def descomponer_en_potencias(valor):
    resultado = []
    for potencia in potencias:
        if valor >= potencia:
            resultado.append(potencia)
            valor -= potencia  # Restamos la potencia del valor
    return resultado

# Función para llenar el DataFrame
def llenar_columna(valor, nombre_columna):
    # Descomponemos el valor en las potencias de 2 correspondientes
    potencias_descompuestas = descomponer_en_potencias(valor)

    # Llenamos el DataFrame según las potencias descompuestas
    for potencia in potencias_descompuestas:
        # Fila 1 corresponde a 2, fila 2 a 4, fila 3 a 8, fila 4 a 16
        if potencia == 2:
            df5.loc[0, nombre_columna] = 'ok'  # Fila 1
        elif potencia == 4:
            df5.loc[1, nombre_columna] = 'ok'  # Fila 2
        elif potencia == 8:
            df5.loc[2, nombre_columna] = 'ok'  # Fila 3
        elif potencia == 16:
            df5.loc[3, nombre_columna] = 'ok'  # Fila 4

# Recorrer la lista de valores binarios y llenar las columnas
for idx, valor in enumerate(segunda_columna_lista):
    nombre_columna = grupo_vals[idx]  # Usamos el nombre de la columna directamente
    llenar_columna(int(valor), nombre_columna)  # Llamamos a la función con el valor y el nombre de la columna

# Remover cualquier fila extra (con NaN) que haya sido creada accidentalmente
df5.dropna(how='all', inplace=True)

# Mostrar el DataFrame resultante
df5
```
